```python
Data=pd.read_csv('dummyML1.csv')
Data
```

# Preprocess

```python
# impute missing value region category
Data['region_category'].fillna('City', inplace=True)
# Menghapus baris yang memiliki nilai hilang di kolom 'gender'
Data = Data.dropna(subset=['gender'])
# impute missing value
Data['joined_through_referral']= Data['joined_through_referral'].fillna(Data['joined_through_referral'].mode()[0])
Data['medium_of_operation']= Data['medium_of_operation'].fillna(Data['medium_of_operation'].mode()[0])
Data['preferred_offer_types']= Data['preferred_offer_types'].fillna(Data['preferred_offer_types'].mode()[0])
```
